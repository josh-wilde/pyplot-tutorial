```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# plt at its simplest takes in a y series
# Will autocreate the x series array starting at 0
plt.plot([1,2,3,4])
```

```python
# Can also specify a particular x series to use
# Third argument is the line style, 'b-' is the default
# Can use the axis command to set axis with this argument:
#[xmin,xmax,ymin,ymax]
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()
```

```python
# Plotting multiple series works like this
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
```

```python
# Can set line properties in the plot command
x = np.arange(0.,5.,0.2)
y = x**2
plt.plot(x, y, linewidth=5.0)
```

```python
# But can also change the properties of a single line in a multiple line plot
# Plot returns list of Line2D objects
x = np.arange(0.,5.,0.2)
y1 = x
y2 = x**2

line1, line2 = plt.plot(x,y1,x,y2)
line1.set_antialiased(False)
plt.show()
```

```python
# Can also set multiple properties on a list of lines
lines = plt.plot(x,y1,x,y2)
plt.setp(lines, color='r', linewidth=3.0)
```

```python
# Can get a list of all of of the line properties to set
plt.setp(lines)
```

```python
# Working with multiple figures and axes
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# This is options, figure will be created by default
# But could be used if there are multiple figures involved
plt.figure(1)

# Subplot specifies: numrows, numcols, current figure number
plt.subplot(2,1,1)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
# can add text at arbitrary location with text()
# Then add xlabel(), ylabel(), and title() in particular locations
# There is also a lot more information on using Latex formatting
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Histogram of data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts', fontsize=14, color='red')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(False)
```

```python
# Helpful to be able to annotate a graph at a particular point
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05)
            )
plt.ylim(-2,2)
```
